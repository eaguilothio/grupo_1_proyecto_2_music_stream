```python
# Importar librerias
import os
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from dotenv import load_dotenv
load_dotenv()
import time
import pandas as pd
```

```python
# Autenticación
auth_manager = SpotifyClientCredentials(client_id=os.getenv("SPOTIFY_CLIENT_ID"),
                                       client_secret=os.getenv("SPOTIFY_CLIENT_SECRET"))
sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
datos = sp.search(q="genre:flamenco year:2014", type='track', limit=50, offset=0)
```

```python
tracks = datos['tracks']

print("Items en esta página:", len(tracks['items']))
print("Items totales:", tracks['total'])
```

```python
num_canciones = len(tracks['items'])
print(num_canciones)
```

```python
canciones = []

for cancion in tracks['items']:
    nombre = cancion['name']
    artista = cancion['artists'][0]['name']
    album = cancion['album']['name']
    año = cancion['album']['release_date'][:4]

    canciones.append({
        'cancion': nombre,
        'artista': artista,
        'album': album,
        'año': año
    })
```

```python
canciones_flamenco = pd.DataFrame(canciones)
canciones_flamenco
```
